##**Biblioteca Numpy: Conceitos e Uso**
 **Numpy** é uma biblioteca matemática pra o python. Ele nos permite fazer computação de forma eficiente e eficaz. É melhor do que python regular por causa de suas incríveis capacidades.

 Um array em python é uma estrutura de dados que permite armazenar uma coleção de intens em uma única variável. os intens podem ser de qualquer tipo, como números, strings ou até mesmo outros arrays. Em python, os arrays são geralmente implementados como listas. I "numpy" fornece uma implementação mais avançada de arrays.

É uma função presente na biblioteca numpy que gera um array contendo uma progressão aritimética de números.

```python
arr = np.arange(0, 100, 1,  'int')
print (arr)
```

#Operações sobre matrizes (tensores)

A biblioteca numpy fornece uma série de operações matemáticas para trablhar com matrizes, incluindo adição, sibtração, multiplicação, divisão, etc.




```python
matriz1 = np.array([[1, 2], [3, 4]])
matriz2 = np.array([[5,6], [7,8]])

matriz_soma = matriz1 + matriz2
print(matriz_soma)
```
